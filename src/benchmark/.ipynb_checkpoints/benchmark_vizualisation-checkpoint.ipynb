```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
src_path = "C:\\Users\\jonas\\Google Drive\\NUS\\Dissertation\\src\\benchmark\\results"

file_basename = "benchmark_model_"
```

```python
file_basename + "{}.csv".format(1)
```

```python
data = {}

for i in range(1,7):
    df = pd.read_csv(os.path.join(src_path,file_basename+"{}.csv".format(i)))
    data[i] = {}
    data[i]["FPR"] = df["FPR"]
    data[i]["TPR"] = df["TPR"]
```

```python
plt.figure(figsize=(7,5))
# plt.figure(figsize=(12,6))

for i in range(1,6):
    plt.plot(data[i]["FPR"], data[i]["TPR"], label="Model {}".format(i))

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves of the Test Set")
plt.xlim(0,1)
plt.ylim(0,1)

plt.legend()
plt.savefig("roc_curves.png")
plt.show()
```
