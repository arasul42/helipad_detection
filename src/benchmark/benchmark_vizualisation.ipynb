# Visualize the results of the benchmark by plotting the ROC curve

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
src_path = "C:\\Users\\jonas\\Google Drive\\NUS\\Dissertation\\src\\benchmark\\results"

file_basename = "benchmark_model_"
```

```python
file_basename + "{}.csv".format(1)
```

```python
data = {}

model_numbers = [1,2,3,4,5,8]

for i in model_numbers:
    df = pd.read_csv(os.path.join(src_path,file_basename+"{}.csv".format(i)))
    data[i] = {}
    data[i]["FPR"] = df["FPR"]
    data[i]["TPR"] = df["TPR"]
```

```python
plt.figure(figsize=(7,5))
# plt.figure(figsize=(12,6))

letters = ['A', 'B', 'C', 'D', 'E', 'F', 'F', 'F', 'F', 'F']
for i in model_numbers:
    plt.plot(data[i]["FPR"], data[i]["TPR"], label="Model {}".format(letters[i-1]))

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
# plt.title("ROC Curves of the Test Set")
plt.xlim(0,0.5)
plt.ylim(0.5,1)

plt.legend()
plt.savefig("roc_curves.png")
plt.show()
```
