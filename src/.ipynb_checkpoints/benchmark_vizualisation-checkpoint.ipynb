```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
src_path = "C:\\Users\\jonas\\Google Drive\\NUS\\Dissertation\\src"

file_basename = "benchmark_model_"
```

```python
file_basename + "{}.csv".format(1)
```

```python
data = {}

for i in range(1,5):
    df = pd.read_csv(file_basename+"{}.csv".format(i))
    data[i] = {}
    data[i]["FPR"] = df["FPR"]
    data[i]["TPR"] = df["TPR"]
```

```python
plt.figure(figsize=(7,5))

for i in range(1,5):
    plt.plot(data[i]["FPR"], data[i]["TPR"], label="Model {}".format(i))

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.savefig("roc_curves.png")
plt.show()
```
